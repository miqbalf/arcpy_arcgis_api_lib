```python
import pandas as pd
import geopandas as gpd
from arcgis.features import GeoAccessor

def remove_suffix(df):
    cols_to_drop = [col for col in df.columns if col.endswith('_x')]
    df = df.drop(columns=cols_to_drop)
    df.columns = [col.replace('_y', '') if col.endswith('_y') else col for col in df.columns]
    return df

# def matching_country_office(input_df_revised, regional_border, world_df,
#                              list_office_id_regional, list_office_id_non_regional,
#                              office_ref_df_regional, office_ref_df_no_regional):
def matching_country_office(input_df_revised,
                             list_office_id_regional, list_office_id_non_regional,
                             office_ref_df_regional, office_ref_df_no_regional):



    # Defensive copies
    input_df_revised = input_df_revised.copy()
    # regional_border = regional_border.copy()
    # world_df = world_df.copy()

    # Split by regional vs non-regional
    input_df_revised_non_regional = input_df_revised.loc[input_df_revised['office'].isin(list_office_id_non_regional)]
    input_df_revised_regional = input_df_revised.loc[input_df_revised['office'].isin(list_office_id_regional)]

    # Join attributes
    df_data_country_joined_nonregional = pd.merge(
        input_df_revised_non_regional, office_ref_df_no_regional,
        left_on='office', right_on='office_gsheet_andreas', how='left',
        suffixes=('_x', '_y')
    )
    df_data_country_joined_regional = input_df_revised_regional.copy()
    df_data_country_joined_regional['country_gsheet_andreas'] = df_data_country_joined_regional['region']
    df_data_country_joined_regional['group_dataset'] = 'regional_grouped'

    df_data_country_joined_nonregional = remove_suffix(df_data_country_joined_nonregional)
    df_data_country_joined_regional = remove_suffix(df_data_country_joined_regional)

    df_data_country_joined_nonregional['country'] = df_data_country_joined_nonregional['country_gsheet_andreas']
    df_data_country_joined_regional['country'] = df_data_country_joined_regional['country_gsheet_andreas']

    # # Geometry sources (expect Shapely in 'SHAPE')
    # regional_border_use = regional_border[['office', 'SHAPE']].copy()
    # world_df_use = world_df[['name', 'SHAPE']].copy()

    # # Attach geometry (still Shapely at this point)
    # gdf_npo_regional = pd.merge(df_data_country_joined_regional, regional_border_use, on='office', how='left')
    # gdf_npo_non_regional = pd.merge(
    #     df_data_country_joined_nonregional, world_df_use,
    #     left_on='country_opensource_data_name', right_on='name', how='left'
    # )

    # # Combine
    # all_npo = pd.concat([gdf_npo_regional, gdf_npo_non_regional], ignore_index=True)
    all_npo = pd.concat([df_data_country_joined_regional,df_data_country_joined_nonregional ], ignore_index=True)
    for c in ('name', 'name_x', 'name_y'):
        if c in all_npo.columns:
            all_npo = all_npo.drop(columns=c)
    # #
    # # gdf = gpd.GeoDataFrame(all_npo, geometry='SHAPE', crs='EPSG:3857')
    # # gdf.to_crs(epsg=3857)
    # # gdf.crs
    # gdf = all_npo.copy()

    # # sdf = GeoAccessor.from_geodataframe(gdf,column_name="SHAPE")
    # sdf = gdf
    # sdf.spatial.set_geometry('SHAPE')
    # print(sdf.spatial.sr)
    # # sdf.spatial.sr

    # return sdf
    return all_npo
```

```python
#### NOW LETS TESTED TO IMPORT BACK TO DF FROM FEATURE CLASS
# function to convert feature into df
def layer_gdf(layer):
    desc = arcpy.Describe(layer)
    field_names = [field.name for field in desc.fields if field.type != 'Geometry']
    
    # Add OID field to have a unique identifier if needed
    if desc.hasOID and desc.OIDFieldName not in field_names:
        field_names.insert(0, desc.OIDFieldName)

    # 2. Use a SearchCursor with the 'SHAPE@' token to get geometry
    features = []
    # The list of fields now also includes 'SHAPE@' to access geometry objects
    with arcpy.da.SearchCursor(layer, field_names + ['SHAPE@']) as cursor:
        for row in cursor:
            # The geometry object is the last item in the row
            geometry = row[-1]
            attributes = dict(zip(field_names, row[:-1]))
            
            # Use the __geo_interface__ to create a GeoJSON-like feature
            if geometry:
                 features.append({
                    'type': 'Feature',
                    'geometry': geometry.__geo_interface__,
                    'properties': attributes
                })

    # 3. Create the GeoDataFrame from the list of features
    gdf = gpd.GeoDataFrame.from_features(features)

    # 4. Set the Coordinate Reference System (CRS)
    #    This step is CRITICAL for any spatial analysis.
    sr = desc.spatialReference
    if sr.factoryCode: # Check if it has an EPSG code
        gdf.set_crs(epsg=sr.factoryCode, inplace=True)
    elif sr.wkt: # Otherwise, use the WKT representation
        gdf.set_crs(crs=sr.wkt, inplace=True)

    return gdf

def layer_df(layer):
    fields = [field.name for field in arcpy.ListFields(layer)]

    # Initialize an empty list to hold the data
    data = []
    
    # Use a SearchCursor to iterate over the rows in the attribute table
    with arcpy.da.SearchCursor(layer, fields) as cursor:
        for row in cursor:
            data.append(row)
    
    # Create a pandas DataFrame from the data
    df = pd.DataFrame(data, columns=fields)
    return df


## implementation
test_df = layer_df(list_full_path_fc[0])
```

```python
import os
import arcpy

def joining_point_one_many(in_table, in_fc_geom, out_fc):
    print(f"\n▶️ Joining points for: {os.path.basename(in_table)}")
   
    in_field, join_field, geom = 'office', 'office', in_fc_geom # we will just have the same for both of them

    # Step 3 — Make feature layer FIRST (so join attaches correctly)
    geom_layer = f"lyr_{os.path.basename(geom)}"
    if arcpy.Exists(geom_layer):
        arcpy.management.Delete(geom_layer)
    arcpy.management.MakeFeatureLayer(geom, geom_layer)
    print(f"  - Created temp feature layer: {geom_layer}")

    # Step 4 — Add join
    arcpy.management.AddJoin(
        in_layer_or_view=geom_layer,
        in_field=in_field,
        join_table=in_table,
        join_field=join_field,
        join_type="KEEP_ALL",
        index_join_fields="INDEX_JOIN_FIELDS",
        rebuild_index="REBUILD_INDEX",
        join_operation="JOIN_ONE_TO_MANY"
    )
    print(f"  - Joined {geom} to {in_table} on {in_field} = {join_field}")

    # Step 5 — Make filtered layer safely (ArcGIS auto-resolves prefixes)
    filtered_layer = f"{geom_layer}_filtered"
    where_clause = "metric_code IS NOT NULL"
    print(f"  - Filtering joined layer with: {where_clause}")

    arcpy.management.MakeFeatureLayer(
        in_features=geom_layer,
        out_layer=filtered_layer,
        where_clause=where_clause
    )

    # Step 6 — Export to output feature class
    print(f"  - Exporting result to: {out_fc}")
    arcpy.conversion.ExportFeatures(filtered_layer, out_fc)

    # Step 7 — Clean up layers
    arcpy.management.Delete(filtered_layer)
    arcpy.management.Delete(geom_layer)

    print(f"✅ Done: Output saved to {out_fc}")


if is_first_run:
    # === Loop ===
    list_fc = []
    
    for i, in_table in enumerate(list_full_path_fc, start=1):
        print(f"\n=== [{i}/{len(list_full_path_fc)}] Processing {os.path.basename(in_table)} ===")
    
        # ALL
        out_fc_all = os.path.join(main_gdb, os.path.basename(in_table) + '_cleaned')
        joining_point_one_many(
            in_table=in_table,
            in_fc_geom=input_fc_geom,
            out_fc=out_fc_all
        )
        list_fc.append(out_fc_all)
    
    print("\n✅ All joins finished successfully.")

else:
    # === Loop ===
    list_fc = []
    
    for i, in_table in enumerate(list_full_path_fc, start=1):
        print(f"\n=== [{i}/{len(list_full_path_fc)}] Processing {os.path.basename(in_table)} ===")
    
        # ALL
        out_fc_all = os.path.join(main_gdb, os.path.basename(in_table) + '_cleaned')
        # joining_point_one_many(
        #     in_table=in_table,
        #     in_fc_geom=input_fc_geom,
        #     out_fc=out_fc_regional
        # )
        list_fc.append(out_fc_all)
    
    print("\n✅ All joins finished successfully.")
    
```

```python
from arcgis.features import GeoAccessor, GeoSeriesAccessor
import math

def append_sdf_to_feature_layer(feature_layer, sdf, batch_size=400):
    """
    Appends data from a Spatially Enabled DataFrame (sdf)
    to a target ArcGIS Feature Layer, in safe batches.

    Args:
        feature_layer: arcgis.features.FeatureLayer
            Target layer to upload features to.
        sdf: Spatially Enabled DataFrame
            Must have a geometry column named 'SHAPE'.
        batch_size: int, optional
            Number of records per batch (default=400).
    """
    # --- Ensure spatial geometry is active ---
    if not hasattr(sdf, "spatial"):
        sdf.spatial.set_geometry("SHAPE")
    else:
        # just make sure geometry is set properly
        if sdf.spatial.name != "SHAPE":
            sdf.spatial.set_geometry("SHAPE")

    # --- Confirm valid geometries ---
    if sdf["SHAPE"].isnull().any():
        raise ValueError("Some records have NULL geometry — check 'SHAPE' column before upload.")

    # --- Convert to features ---
    features = sdf.spatial.to_featureset().features

    # --- Split into batches ---
    total = len(features)
    n_batches = math.ceil(total / batch_size)

    print(f"📦 Preparing to upload {total} features in {n_batches} batches...")

    # --- Upload loop ---
    success, failed = 0, 0
    for i in range(n_batches):
        start = i * batch_size
        end = min(start + batch_size, total)
        batch = features[start:end]

        try:
            res = feature_layer.edit_features(adds=batch)
            adds = res.get("addResults", [])
            if all(r.get("success") for r in adds):
                success += len(adds)
                print(f"✅ Batch {i+1}/{n_batches} uploaded ({len(adds)} features).")
            else:
                failed += len(adds)
                print(f"⚠️ Batch {i+1}/{n_batches} failed partially.")
        except Exception as e:
            print(f"❌ Batch {i+1}/{n_batches} failed: {e}")
            failed += len(batch)

    print(f"\n🎯 Upload complete: {success} succeeded, {failed} failed.")
```
